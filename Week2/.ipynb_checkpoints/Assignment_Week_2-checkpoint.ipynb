# Excercise 2

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
Auto = pd.read_csv("./data/Auto.csv")
Auto.head()
```

### Q1.

```python
Auto["horsepower"] = pd.to_numeric(Auto["horsepower"], errors="coerce")
Auto = Auto.dropna()

X = Auto["horsepower"]
y = Auto["mpg"]

X = sm.add_constant(X)  # adds intercept
model = sm.OLS(y, X).fit()

model.summary()
```

### (a) Relationship between predictor and response
The horsepower coefficient is negative and highly statistically significant,
indicating that higher horsepower is associated with lower fuel efficiency.

### (b) Prediction at horsepower = 98

```python
X_new = pd.DataFrame({"const": [1], "horsepower": [98]})

prediction = model.get_prediction(X_new)
prediction.summary_frame(alpha=0.05)
```

The predicted mpg at horsepower = 98 is approximately 24 mpg.
The confidence interval is narrower than the prediction interval
because it reflects uncertainty in the mean response.

## Q2. 

```python
plt.figure()
plt.scatter(Auto["horsepower"], Auto["mpg"])
plt.plot(Auto["horsepower"], model.predict(X), color="red")
plt.xlabel("Horsepower")
plt.ylabel("Miles Per Gallon")
plt.show()
```

### Q3.

```python
# Residuals vs Fitted
plt.figure()
plt.scatter(model.fittedvalues, model.resid)
plt.axhline(0)
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted")
plt.show()
```

```python
# Normal Q-Q Plot
sm.qqplot(model.resid, line="45")
plt.title("Normal Q-Q Plot")
plt.show()
```

```python
# Scale-Location Plot
plt.figure()
plt.scatter(model.fittedvalues, (abs(model.resid))**0.5)
plt.xlabel("Fitted Values")
plt.ylabel("|Residuals|")
plt.title("Scale-Location Plot")
plt.show()
```

```python
# Residuals vs Leverage
sm.graphics.plot_leverage_resid2(model)
plt.show()
```

Regressions assumptions dont apply as these graphs indicate that the relationship between horsepower and mpg is not perfectly linear

# Exercise 3
### Q1.

```python
data = pd.read_csv("./data/inputData_AUDUSD_20120426.csv")
data.head()
```

```python
data["Date"] = pd.to_datetime(data["Date"])
```

```python
plt.figure(figsize=(12, 5))
plt.plot(data["Date"], data["Close"])
plt.xlabel("Date")
plt.ylabel("USD/CAD Closing Price")
plt.title("Daily Closing Price of USD/CAD")
plt.show()
```

Its a random walk graph and it has no clear mean that can be inferred

### Q2.

```python
from statsmodels.tsa.stattools import adfuller
data["log_price"] = np.log(data["Close"])
data[["Date", "log_price"]].head()
```

```python
adf_result = adfuller(data["log_price"], maxlag=1)
t_statistic = adf_result[0]
p_value = adf_result[1]
critical_values = adf_result[4]

print("ADF t-statistic:", t_statistic)
print("p-value:", p_value)
print("90% critical value:", critical_values["10%"])
```

The Augmented DickeyFuller test gave us a test statistic that is higher than the 90% critical value and a p-value of 0.42(approx) which is greater than 0.10. Therefore, we fail to reject the null hypothesis of a unit root, indicating that the log-price series is **not stationary** 

### Q3.

```python
y = np.log(data["Close"])
delta_y = y.diff().dropna()
y_lag = y.shift(1).dropna()
delta_y = delta_y.loc[y_lag.index]
```

```python
X = sm.add_constant(y_lag)   # adds 
model_hl = sm.OLS(delta_y, X).fit()

print(model_hl.summary())
```

```python
lambda_hat = model_hl.params.iloc[1]

half_life = np.log(2) / (-lambda_hat)

print("Estimated lambda:", lambda_hat)
print("Half-life (days):", half_life)
```

The  half-life of mean reversion is around 167 days (167.42 to be precise in terms of value).
It isnt practical at all for short term traders due to its high magnitude which is in the order of a few months

# Exercise 4

### Q1.

```python
data = pd.read_csv("./data/inputData_EWA_EWC.csv")
data.head()
```

```python
data["Date"] = pd.to_datetime(data["Date"])
plt.figure(figsize=(10, 5))
plt.plot(data["Date"], data["EWA"], label="EWA (Australia)")
plt.plot(data["Date"], data["EWC"], label="EWC (Canada)")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Daily Closing Prices of EWA and EWC")
plt.legend()
plt.show()
```

```python
correlation = data["EWA"].corr(data["EWC"])
print("Correlation coefficient:", correlation)
```

EWA and EWC show similar movements over time. 
The correlation coefficient between their price series is around 0.95 indicating a strong positive relationship

### Q2.

```python
Y = data["EWA"]
X = data["EWC"]     
```

```python
X = sm.add_constant(X)
```

```python
ols_model = sm.OLS(Y, X).fit()
print(ols_model.summary())
```

```python
hedge_ratio = ols_model.params.iloc[1]
print("Hedge Ratio ():", hedge_ratio)
```

### Q3.

```python
data["Spread"] = data["EWA"] - hedge_ratio * data["EWC"]
data[["Date", "Spread"]].head()
```

### Q4.

```python
adf_result = adfuller(
    data["Spread"].dropna(),
    maxlag=1,
    regression="n"
)
```

```python
t_statistic = adf_result[0]
p_value = adf_result[1]
critical_values = adf_result[4]

print("ADF t-statistic:", t_statistic)
print("p-value:", p_value)
print("95% critical value:", critical_values["5%"])
```

### Q5.

```python
spread = data["Spread"]
```

```python
delta_spread = spread.diff().dropna()
spread_lag = spread.shift(1).dropna()
delta_spread = delta_spread.loc[spread_lag.index]
```

```python
X = spread_lag.values.reshape(-1, 1)
model_hl = sm.OLS(delta_spread, X).fit()

print(model_hl.summary())
```

```python
lambda_hat = model_hl.params.iloc[0]

half_life = np.log(2) / (-lambda_hat)

print("Estimated lambda:", lambda_hat)
print("Half-life (days):", half_life)
```

The half life is around 288 days
